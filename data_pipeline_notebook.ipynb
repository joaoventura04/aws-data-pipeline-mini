```python

import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files


print("Selecione o ficheiro CSV para upload:")
uploaded = files.upload()
filename = list(uploaded.keys())[0]


df = pd.read_csv(filename)

df['Year'] = df['Year'].astype(int)
df['Mean'] = df['Mean'].astype(float)

plt.figure(figsize=(12,6))
plt.plot(df['Year'], df['Mean'], color='red', marker='o')
plt.xlabel("Year")
plt.ylabel("Mean Temperature")
plt.title("Global Average Temperature")
plt.grid(True)
plt.show()


# Média por século
df['Century'] = df['Year'].apply(lambda x: '1800s' if x < 1900 else '1900s')
mean_by_century = df.groupby('Century')['Mean'].mean()
print("\nMédia por século:")
print(mean_by_century)

# Top 5 anos mais quentes
top5 = df.sort_values('Mean', ascending=False).head(5)
print("\nTop 5 anos mais quentes:")
print(top5)
```
